```python
city     = 'Luxembourg--Luxemburg'
main_url = 'https://www.airbnb.com'
city_url = f'{main_url}/s/{city}/homes/'

# Create selector
html = requests.get(city_url).content
sel  = Selector(text=html)

# Get hotels
hotels = sel.css('div._1e9w8hic')
hotels2 = sel.xpath('//*[@id="site-content"]/div[2]/div/div[7]/div/div/div/div/div/div/div[2]/div/div/div/div/div/div[1]/div/div/div[2]/div/div/div/div[2]/a/@href').extract_first()
print('Number of hotels:', len(hotels))
print('Number of hotels:', len(hotels2))

# Find the next page url
next_page = sel.xpath('//a[contains(@aria-label, "Next")]/@href').extract_first()
print('Next page:', '\n', next_page)
```
